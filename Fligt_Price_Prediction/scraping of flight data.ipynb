```python
flight_name = []

flight_tag = driver.find_elements_by_xpath("//div[@class = 'LegLogo_legImage__MmY0Z']/div/img")
for i in flight_tag[0:55]:
    flight_name.append(i.get_attribute('alt'))
    
print(len(flight_name))
print(flight_name[-1])
```

```python
duration = []
stop_number = []

duration_tag = driver.find_elements_by_xpath("//div[@class='LegInfo_stopsContainer__NWIyN']/span")
for i in duration_tag[0:55]:
    duration.append(i.text)
    
    
stops_tag = driver.find_elements_by_xpath("//div[@class = 'LegInfo_stopsLabelContainer__MmM0Z']")
for i in stops_tag[0:55]:
    stop_number.append(i.text)
    
    
print(len(duration))
print(duration[-1])
print('__________________________')
print(len(stop_number))
print(stop_number[-1])
```

```python
time_tag2 = driver.find_elements_by_xpath("//div[@class='LegInfo_routePartialArrive__Y2U1N']/span/div")
for i in time_tag2[0:55]:
    arrival_time.append(i.text)

print(len(arrival_time))
print(arrival_time[-1])
```

```python
ticket_price = []

price_tag = driver.find_elements_by_xpath("//div[@class='TicketStub_horizontalStubContainer__YjM5M']/div/div")
for i in price_tag[0:55]:
    ticket_price.append(i.text)

print(len(ticket_price))
print(ticket_price[-1])
```
