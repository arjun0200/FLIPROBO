### 8. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)
driver.get('https://www.forbes.com/billionaires/')
driver.maximize_window()
```

```python
Ranks = []
Name = []
NetWorth = []
Age = []
citizen = []
category = []
```

```python
rank = driver.find_elements_by_xpath("//div[@class = 'rank']")
name = driver.find_elements_by_xpath("//div[@class = 'personName']")
netWorth = driver.find_elements_by_xpath("//div[@class = 'netWorth']")
age = driver.find_elements_by_xpath("//div[@class = 'age']/div")
citizenship = driver.find_elements_by_xpath("//div[@class = 'countryOfCitizenship']")
categories = driver.find_elements_by_xpath("//div[@class = 'category']")

for rank, name, netWorth, age, citizenship, categories in zip(rank, name, netWorth, age, citizenship, categories):
    Ranks.append(rank.text)
    Name.append(name.text)
    NetWorth.append(netWorth.text)
    Age.append(age.text)
    citizen.append(citizenship.text)
    category.append(categories.text)
```

```python
df = pd.DataFrame({'Ranks': Ranks, 'Name': Name, 'NetWorth': NetWorth, 'Age': Age, 'Citizenship': citizen, 'Technology': category})
df
```

### 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted  from any YouTube Video.

```python
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)
driver.get('https://www.youtube.com/watch?v=Di1TbA7MHUs')
driver.maximize_window()
```

```python
comments = []
votes = []
posted = []


#comments
com = driver.find_elements_by_xpath("//yt-formatted-string[@id = 'content-text']")
for i in com:
    try:
        comments.append(i.text)
    except NoSuchElementException:
        comments.append('-')
        
#upvotes counts
vote = driver.find_elements_by_xpath("//span[@id = 'vote-count-middle']")
for i in vote:
    try:
        votes.append(i.text)
    except NoSuchElementException:
        votes.append("-")
        

#posted comment time    
posted_comment = driver.find_elements_by_xpath("//yt-formatted-string[@class = 'published-time-text above-comment style-scope ytd-comment-renderer']")
for i in posted_comment:
    try:
        posted.append(i.text)
    except NoSuchElementException:
        posted.append('_')
```

```python
df = pd.DataFrame({'Comments': comments, 'Up Votes': votes, 'Posted Date': posted})
```
