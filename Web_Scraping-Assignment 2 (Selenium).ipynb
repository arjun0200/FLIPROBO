```python
driver4 = webdriver.Chrome()
```

```python
driver4.get('https://www.flipkart.com/')
time.sleep(2)
driver4.maximize_window()
```

```python
search4 = driver4.find_element_by_xpath("//input[@class='_3704LK']")
search4.send_keys("sunglasses")
time.sleep(2)
search4.send_keys(Keys.ENTER)
```

```python
#creating list
brand_names = []
product_des = []
sunglasses = []
```

```python
for page in range(1,4,1):
    
    page_url4 = "https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=" + str(page)
    driver4.get(page_url4)
    
    #creating iterables
    
    brand = driver4.find_elements_by_xpath("//div[@class='_2WkVRV']")
    product_desc = driver4.find_elements_by_xpath("//a[@class='IRpwTa']")
    glasses_price = driver4.find_elements_by_xpath("//div[@class='_30jeq3']")
    
    for brand, product_desc, glasses_price in zip(brand, product_desc, glasses_price):
        brand_names.append(brand.text)
        product_des.append(product_desc.text)
        sunglasses.append(glasses_price.text)
        
        
#print(len(brand_names),len(product_des), len(sunglasses))
#120 120 120    
```

```python
Sunglasses = pd.DataFrame()
Sunglasses['Brand Names'] = brand_names[0:100]
Sunglasses['Product Description'] = product_des[0:100]
Sunglasses['Price'] = sunglasses[0:100]
```

```python
pd.set_option("display.max_rows", None)
print("----------------------------------------100 Sunglasses----------------------------------------------")
Sunglasses
```

```python
driver4.close()
```

```python
#searching i7 laptops by filtering
search_i7 = driver8.find_element_by_xpath("//span[text()='Intel Core i7']").click()
```

```python
#searching i9 laptops by filtering
#search_i9 = driver8.find_element_by_xpath("//span[text()='Intel Core i9']").click()
```

```python
#scraping name of the laptops
laptop_name = []

laptops = driver8.find_elements_by_xpath("//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']/span[1]")
for i in laptops:
    laptop_name.append(i.text)
```

```python
# scraping prices of the laptops
prices = []

price = driver8.find_elements_by_xpath("//span[@class='a-price-whole']")
for i in price:
    prices.append(i.text.replace(",",""))
```

```python
# scraping ratings of the laptops
ratings = []
ratings_box = driver8.find_elements_by_xpath('.//div[@class="a-row a-size-small"]/span')
for i in ratings_box:
        ratings.append(i.get_attribute('aria-label'))
        if i.get_attribute('aria-label') == np.NaN:
            ratings.append('-')
```

```python
laptop_details = pd.DataFrame()
laptop_details['Laptops'] = laptop_name[0:10]
laptop_details['Prices'] = prices[0:10]
laptop_details['Ratings'] = ratings[0:20:2]
```

```python
print("----------------------------------------First 10 Laptops Details----------------------------------------------")
laptop_details
```
