1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
driver1 = webdriver.Chrome("chromedriver.exe")
time.sleep(2)
driver1.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver1.maximize_window()
time.sleep(2)

delay = 10
WebDriverWait(driver1, delay).until(EC.presence_of_element_located((By.CLASS_NAME, 'reference')))

ranks = []
views = []
dates = []
video_name = []
singer = []

table = driver1.find_elements_by_xpath('//table[@class = "wikitable sortable jquery-tablesorter"]/tbody/tr/td') 
rank = driver1.find_elements_by_xpath('//td[@align = "center"][1]')  #extracting ranks
view = driver1.find_elements_by_xpath('//td[@align = "center"][2]')  #extracting views
date = driver1.find_elements_by_xpath('//td[@align = "right"]')     #extracting dates

for i in table[1:180:6]:
    video_name.append(i.text)

for j in table[2:180:6]:
    singer.append(j.text)
    
for k in rank:
    ranks.append(k.text)

for l in view:
    views.append(l.text)
    
for m in date:
    dates.append(m.text)
```

```python
print(len(ranks[:30]))
print(len(views[:30]))
print(len(dates[:30]))
print(len(video_name))
print(len(singer))
```

```python
#Making DataFrame
df = pd.DataFrame({"Ranks": ranks[:30], "Video_Name": video_name, "Singer": singer, "Views_in_Millions": views[:30], "Released_Dates": dates[:30]})
df
```

```python
driver1.close()
```

6. Scrape the details of top 100 songs on billiboard.com.
Url = https:/www.billboard.com/
You have to find the following details:
    
A) Song name

B) Artist name

C) Last week rank

D) Peak rank

E) Weeks on board

Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
driver6 = webdriver.Chrome("chromedriver.exe")
time.sleep(2)
driver6.get('https://www.billboard.com/')
driver6.maximize_window()
time.sleep(2)
delay = 20
WebDriverWait(driver6, delay).until(EC.presence_of_element_located((By.CLASS_NAME, 'admz')))
```

```python
tag = driver6.find_elements_by_xpath("//a[@class = 'c-link  lrv-a-unstyle-link lrv-u-color-brand-primary:hover lrv-a-hover-effect lrv-u-whitespace-nowrap lrv-u-color-grey-dark']")
for i in tag:
    if i.text =='Hot 100':
        url = i.get_attribute('href')
        
driver6.get(url)
```

```python
rank_tag = driver6.find_elements_by_xpath("//span[@class = 'c-label  a-font-primary-bold-l u-font-size-32@tablet u-letter-spacing-0080@tablet']")
song_name_tag = driver6.find_elements_by_xpath("//li[@class = 'lrv-u-width-100p']/ul/li/h3")

rank = []
song_name = []
peak_pos = []
artist = []
last_week = []
wks_of_chart = []

#-----------------------------rank and song name -----------------------------------
try :    
    for i, song in zip(rank_tag, song_name_tag):
        rank.append(int(i.text))
        song_name.append(song.text)
except:
    rank.append('-')
    song_name.append('-')
        
details_tag = driver6.find_elements_by_xpath("//li[@class = 'lrv-u-width-100p']/ul/li/span")
details_tag[0].text

# --------------------------fetching artist-----------------------------
try:
    for i in details_tag[0::4]:
        artist.append(i.text)
except:
    artist.append('-')

    
# --------------------------- last week ---------------------------------
try:
    for i in details_tag[1::4]:
        last_week.append(i.text)
except:
    last_week.append('-')
    
# -------------------------------peak pos -----------------------------------
try:
    for i in details_tag[2::4]:
        peak_pos.append(i.text)
except:
    peak_pos.append('-')
    
# ------------------------------wks_of_chart ------------------------------

try:
    for i in details_tag[3::4]:
        wks_of_chart.append(i.text)
except:
    wks_of_chart.append('-')
```

```python
df6 = pd.DataFrame()
df6['Rank'] = rank
df6['Song Name'] = song_name
df6['Artist'] = artist
df6['Last Week Rank'] = last_week
df6['Peak Rank'] = peak_pos
df6['Weeks on Board'] = wks_of_chart
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df6
```

10. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/
You have to find the following details:
    
A) Dataset name

B) Data type

C) Task

D) Attribute type

E) No of instances

F) No of attribute

G) Year

Note: - from the home page you have to go to the ShowAllDataset page through code

```python
driver10 = webdriver.Chrome("chromedriver.exe")
time.sleep(2)
driver10.get('https://archive.ics.uci.edu/ml/index.php')
driver10.maximize_window()
time.sleep(10)
```

```python
search = driver10.find_element_by_xpath("//span[@class = 'normal']/b/a")
driver10.get(search.get_attribute('href'))
```

```python
dataset = []  # Dataset names
dataset_url = driver10.find_elements_by_xpath("//p[@class = 'normal']/b/a")
for i in dataset_url:
    dataset.append(i.text)
```

```python
url = []  # dataset urls
for i in dataset_url:
    url.append(i.get_attribute("href"))
```

```python
task = []
data_type = []
attribute_type = []
instances = []
attribute = []
year = []


for j in url:
    driver10.get(j)
    time.sleep(2)
    # --------------------------------Attribute type ----------------------------------
    try:
        attribute_type_tag = driver10.find_elements_by_xpath("//table[@border = 1]/tbody/tr/td/p")
        attribute_type.append(attribute_type_tag[7].text)

    except:
        attribute_type.append('-')
    
    # --------------------------------data Type --------------------------------------
    try:  
        data_type_tag = driver10.find_elements_by_xpath("//table[@border = 1]/tbody/tr/td/p")
        data_type.append(data_type_tag[1].text)

    except:
        data_type.append('-')

    #  ----------------------------------------- Task --------------------------------------
    try:  
        task_tag = driver10.find_elements_by_xpath("//table[@border = 1]/tbody/tr/td/p")
        task.append(task_tag[13].text)

    except:
        task.append('-')

    # ------------------------------------------ Instances --------------------------------------
    try:  
        instances_tag = driver10.find_elements_by_xpath("//table[@border = 1]/tbody/tr/td/p")
        instances.append(instances_tag[3].text)

    except:
        instances.append('-')
    
       # ------------------------------------ No. of Attributes  --------------------------
    
    try:  
        attribute_tag = driver10.find_elements_by_xpath("//table[@border = 1]/tbody/tr/td/p")
        attribute.append(attribute_tag[9].text)

    except:
        attribute.append('-')
        
        
       # ------------------------------------year   ------------------------------------
    try:  
        year_tag = driver10.find_elements_by_xpath("//table[@border = 1]/tbody/tr/td/p")
        year.append(year_tag[11].text.split('-')[0])

    except:
        year.append('-')
```

```python
df10 = pd.DataFrame()
df10['Dataset Name'] =  dataset[:100]
df10['Data Type'] = data_type[:100]
df10['Task'] =  task[:100]
df10['Attribute Type'] =  attribute_type[:100]
df10['No. of Instances'] = instances[:100]
df10['No. of Attributes'] =   attribute[:100]
df10['Year'] =  year[:100]
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df10
```
