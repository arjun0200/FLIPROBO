### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
url1 = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
page1 = requests.get(url1)
page1   #for checking page response
```

```python
# creating list
Movies = []  
Ratings = []

soup=BeautifulSoup(page1.content)

movie_name = soup.find_all('td', class_ = "titleColumn")
ratings = soup.find_all('td', class_ = "ratingColumn imdbRating")

for i in range(0,100):   # for 100 movies data
    Movies.append(movie_name[i].text.split("\n"))
    Ratings.append(ratings[i].text.split("\n"))

#Making DataFrame

df = pd.DataFrame(Movies, columns = ['1','Sr.', 'Movie_names', 'Year_of_Release','4'])
Ratings = pd.DataFrame(Ratings)
df['Imdb Ratings'] = Ratings.iloc[:,1]
Imdb_movies = df.drop(['1','4'], axis=1)
print("-------------------- Imdb Movies----------------------- ")
Imdb_movies       # Making a dataframe
```

### 7 ) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post

```python
#creating list
heading7 = []
date7 = []
content7 = []
Video_url7 = []


heading = soup7.find_all("h2", class_ = "entry-title")
for i in heading:
    heading7.append(i.text)
    
date = soup7.find_all("time", class_ = "entry-time")
for i in date:
    date7.append(i.text.strip())

    
content = soup7.find_all("div", class_ = "entry-content")
for i in content:
    content7.append(i.text.replace("\n", ""))

for i in soup7.find_all('article'):
    Video_url7.append(soup7.find("iframe", class_ = "youtube-player")['src'])
```

```python
#creating DataFrame

Coreyms = pd.DataFrame({"Heading": heading7, "Posted_Date": date7, "Content": content7, "Video_URL": Video_url7})
Coreyms
```

### 9) Write a python program to scrape mentioned details from dineout.co.in :i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL 

```python
url9 = "https://www.dineout.co.in/delhi-restaurants/buffet-special"
URL9 = requests.get(url9)
URL9                     #for checking page response
```

```python
soup9 = BeautifulSoup(URL9.content)
#print(soup9.prettify())
```

```python
# Creating list

Restaurant_name = []
Cuisine = []
Location = []
Ratings = []
Image = []

# Appending list data

for i in soup9.find_all("a", class_ = "restnt-name ellipsis"):
    Restaurant_name.append(i.text.strip())
    
for i in soup9.find_all('span', class_ = 'double-line-ellipsis'):
    Cuisine.append(i.text)

for i in soup9.find_all("div", class_ = "restnt-loc ellipsis"):
    Location.append(i.text.strip())   
   
for i in soup9.find_all("div", class_ = "restnt-rating rating-4"):
    Ratings.append(float(i.text.strip()))
    
for i in soup9.find_all("img", class_ = "no-img"):
    Image.append(i["data-src"])
    
```

```python
df = pd.DataFrame(Cuisine ,columns=["name"])
df1 =df['name'].str.split('|', expand=True)
Cuisine1 = df1.iloc[:,1:2]
```

```python
# -----------checkling lengths----------
    
# print(len(Restaurant_name))
# print(len(Cuisine1))
# print(len(Location))
# print(len(Ratings))
# print(len(Image))


# ----------Creating DataFrame-------
Res = pd.DataFrame()
Res["Restaurant_Name"] = Restaurant_name
Res["Cuisine"] = Cuisine1
Res["Location"] = Location
Res["Ratings"] = Ratings
Res["Images_URL"] = Image
```
